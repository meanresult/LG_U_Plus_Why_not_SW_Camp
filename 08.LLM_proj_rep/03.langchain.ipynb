# 처리방식 

## 처리 메소드
invoke()

`invoke()` 메서드는 주어진 입력을 처리하고 단일 응답을 반환합니다. 이 메서드는 동기적으로 동작하며, 입력 메시지를 처리한 후 결과를 반환합니다.

batch() 

중복되는 토큰들을 줄여서 묶어 처리하는 방식

`batch()` 메서드는 여러 개의 입력을 한 번에 처리하여 각각의 입력에 대한 응답을 반환합니다. 이 메서드는 동기적으로 동작하며, 입력 리스트에 대한 응답 리스트를 반환합니다.

stream()

`stream()` 메서드는 입력을 처리하면서 실시간으로 응답을 스트리밍합니다. 이 메서드는 비동기적으로 동작하며, 응답을 부분적으로 받을 수 있습니다.

# batch()

```python
message = [
    ("system", "당신은 서울의 음식과 문화 전문가 입니다.")
    ,("human", "노량진을 대표하는 음식을 맛볼 수 있는 레스토랑 5개를 추천해주세요")
]
res1 = llm.batch(message)
res1
```

```python
for re in res1:
    print(re.content)
```

# stream

```python
message = [
    ("system", "당신은 서울의 음식과 문화 전문가 입니다.")
    ,("human", "노량진을 대표하는 음식을 맛볼 수 있는 레스토랑 5개를 추천해주세요")
]
res2 = llm.stream(message)
for re in res2:
    print(re.content, end='')
```

# async, sync

```python
import asyncio
import time

async def invoke_async(llm, messages):
    response = await llm.ainvoke(messages)
    print(response.content)

async def invoke_parallel(llm, messages):
    tasks = [invoke_async(llm, messages) for _ in range(10)]
    await asyncio.gather(*tasks)
```

```python
message = [
    ("system", "당신은 서울의 음식과 문화 전문가 입니다.")
    ,("human", "노량진을 대표하는 음식을 맛볼 수 있는 레스토랑 5개를 추천해주세요")
]

start = time.perf_counter()

await invoke_parallel(llm, message)
end = time.perf_counter()

print(f'Elapsed time: {end - start:.2f} seconds')
```
