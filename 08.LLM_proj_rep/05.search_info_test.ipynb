```python

import yfinance as yf
```

```python
tiker = 'MSFT' # symbol / 심부름 용지 (어디 매장갈지)

msft = yf.Ticker(tiker) # Ticker라는애가 심부름꾼 
msft
```

```python
basic_info = { #해당 매장에서 뭘 가져올지 적어둔 것)
               '회사명'          : msft.info.get('longName')
                ,'산업'          : msft.info.get('industry')
                ,'부문'          : msft.info.get('sector')
                ,'시가총액'       : msft.info.get('marketCap')
                ,'총 발행주식수'  : msft.info.get('shareOutstanding')
                ,'본사 위치'      : msft.info.get('country')
                ,'ceo'           : msft.info.get('ceo')
                ,'웹사이트'       : msft.info.get('website')
}
```

```python
# 기본 정보 출력
for key, value in basic_info.items():
    print(f'{key},{value}')
```

```python
# 재무제표 정보 가져오기 ㅎㅎ 

income_statment = msft.financials # 손익계산서 
balance_shee = msft.balance_sheet #대차대조표 자산 부채 
cashflow_statment = msft.cash_flow # 현금흐름표 ㅎㅎ 
```
