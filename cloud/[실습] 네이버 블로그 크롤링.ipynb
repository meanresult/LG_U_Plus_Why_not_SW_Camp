```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

title_list = []
image_list = []
link_list = []
text_list = []

for page in range(1, 16):
    print(f"\n--- {page}í˜ì´ì§€ ìˆ˜ì§‘ ì¤‘ ---")
    url = f"https://section.blog.naver.com/Search/Post.naver?pageNo={page}&keyword=ë‚¨ìëŠì¢‹ì‚¬ë³µíŒ¨ì…˜"
    
    chrome_options = Options()
    chrome_options.add_experimental_option("detach", False)
    driver = webdriver.Chrome(options=chrome_options)

    try:
        driver.get(url)
        time.sleep(3)

        soup = BeautifulSoup(driver.page_source, 'html.parser')

        title = soup.select("strong.title_post")
        image = soup.select("div.thumbnail_area img")
        link = soup.select("div.desc > a")
        text = soup.select("div.desc .text")

        # âœ… ì¢…ë£Œ ì¡°ê±´: ê²Œì‹œë¬¼ì´ ì—†ì„ ê²½ìš°
        if len(title) == 0:
            print("âŒ ë” ì´ìƒ ìˆ˜ì§‘í•  í¬ìŠ¤íŠ¸ê°€ ì—†ìŠµë‹ˆë‹¤. í¬ë¡¤ë§ì„ ì¢…ë£Œí•©ë‹ˆë‹¤.")
            break

        print(f'âœ… ìˆ˜ì§‘ëœ ë¸”ë¡œê·¸ ìˆ˜: {len(title)}')

        title_list.extend(title)
        image_list.extend(image)
        link_list.extend(link)
        text_list.extend(text)

    finally:
        driver.quit()  # ë¸Œë¼ìš°ì € ê¼­ ë‹«ê¸°
```
