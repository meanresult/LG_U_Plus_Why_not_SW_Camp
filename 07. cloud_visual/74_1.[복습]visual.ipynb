# plt.subplot / plt.subplots ?

✅ 1. plt.subplot()   
도화지에 ‘하나의 작은 그림(축)’을 추가하는 함수   
(nrows, ncols, index) 형식으로 씁니다.   

✅ 2. plt.subplots()    
도화지(Figure)와 축(Axes) 객체를 한 번에 만드는 함수   
객체 지향적으로 그래프를 다룰 때 더 자주 쓰여요.    

```python
import matplotlib.pyplot as plt

# 2행 2열짜리 도화지 중 첫 번째 위치(왼쪽 위)에 그래프 하나 추가
plt.subplot(2, 2, 1)
plt.plot([1, 2, 3], [1, 4, 9])  # 첫 번째 그래프

plt.subplot(2, 2, 4)
plt.plot([1, 2, 3], [1, 2, 3])  # 두 번째 그래프

plt.show()
```

```python
fig, ax = plt.subplots()  # 도화지(fig), 축(ax) 1개 생성
ax.plot([1, 2, 3], [1, 4, 9])  # ax 위에 그래프 그림
plt.show()
```
