```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time
import requests
from PIL import Image
from io import BytesIO
import os
import re

image_dir = 'images'
if not os.path.exists(image_dir):
    os.makedirs(image_dir)

pattern = r'[\\/:"*?<>|]'

for page in range(1, 11):
    print(f"\n--- {page}페이지 수집 중 ---")
    url = f"https://section.blog.naver.com/Search/Post.naver?pageNo={page}&keyword=사복패션"
    
    # 🔄 매번 새로운 브라우저 실행
    chrome_options = Options()
    chrome_options.add_experimental_option("detach", False)
    driver = webdriver.Chrome(options=chrome_options)
    
    try:
        driver.get(url)
        time.sleep(3)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        title_list = soup.select("strong.title_post")
        image_list = soup.select("div.thumbnail_area img")
        link_list = soup.select("div.desc > a")

        for title, image in zip(title_list, image_list):
            image_src = image.get('src')
            if not image_src: continue
            image_url = image_src if image_src.startswith("http") else "https://blog.naver.com" + image_src

            try:
                res = requests.get(image_url, timeout=5)
                img = Image.open(BytesIO(res.content))
                clean_title = re.sub(pattern, '', title.text.strip())
                img_path = os.path.join(image_dir, clean_title + ".png")
                img.save(img_path)
                print(f"✅ 저장: {clean_title}")
            except Exception as e:
                print(f"❌ 저장 실패: {e}")

    except Exception as e:
        print(f"🚨 페이지 오류: {e}")
    
    finally:
        driver.quit()  # 🔒 무조건 브라우저 종료
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
import time

title_list = []
image_list = []
link_list = []
text_list = []

for page in range(1, 16):
    print(f"\n--- {page}페이지 수집 중 ---")
    url = f"https://section.blog.naver.com/Search/Post.naver?pageNo={page}&keyword=남자느좋사복패션"
    
    chrome_options = Options()
    chrome_options.add_experimental_option("detach", False)
    driver = webdriver.Chrome(options=chrome_options)

    try:
        driver.get(url)
        time.sleep(3)

        soup = BeautifulSoup(driver.page_source, 'html.parser')

        title = soup.select("strong.title_post")
        image = soup.select("div.thumbnail_area img")
        link = soup.select("div.desc > a")
        text = soup.select("div.desc .text")

        # ✅ 종료 조건: 게시물이 없을 경우
        if len(title) == 0:
            print("❌ 더 이상 수집할 포스트가 없습니다. 크롤링을 종료합니다.")
            break

        print(f'✅ 수집된 블로그 수: {len(title)}')

        title_list.extend(title)
        image_list.extend(image)
        link_list.extend(link)
        text_list.extend(text)

    finally:
        driver.quit()  # 브라우저 꼭 닫기
```

```python
# 제목, 이미지, 링크를 한 줄씩 같이 다루되, 짧은 쪽에 맞춰서 처리
blog_data = []

for 제목, 이미지, 링크, 텍스트 in zip(title_list, image_list, link_list,text_list):
    blog_data.append({
        'title': 제목.text,
        'contents' : 텍스트.text,
        'link': 링크.get('href'),
        'image': 이미지.get('src')
        
        
    })
```

```python
import pandas as pd

# 리스트 → 표 형태(DataFrame)로 변환
df = pd.DataFrame(blog_data)

# CSV 파일로 저장 (엑셀에서 한글 안 깨지게 하기 위해 utf-8-sig 사용)
df.to_csv("naver_blog_data.csv", encoding="utf-8-sig", index=False)

print("✅ 저장 완료: naver_blog_data.csv")
```
